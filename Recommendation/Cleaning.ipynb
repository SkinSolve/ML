```python
# Fungsi untuk mengganti brand
def replace_brand(skincare_name):
    if pd.isna(skincare_name):
        return skincare_name
    for brand in brand_list:
        if brand.lower() in str(skincare_name).lower():
            return brand
    return skincare_name

# Mengganti brand pada kolom 'skincare_name'
df['brand'] = df['skincare_name'].apply(replace_brand)

# Menampilkan lima baris pertama
print(df.head())
```
